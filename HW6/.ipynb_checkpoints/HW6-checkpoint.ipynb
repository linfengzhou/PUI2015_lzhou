#Task 1: 
##Event detection: Identify the most prominent event. There is a very significant drop (>3-sigma) in all time series.
##Identify it and figure out what it is due to.

```python
mta = data
print (mta.shape)
mta_allsum = np.nansum(mta, axis=0)
print (mta_allsum.shape)
#or for all ride types
mta_allsum_1 = np.nansum(mta, axis=1)
print (mta_allsum_1.shape)
```

```python
df = []
station = {'swpies':mta_allsum[0].reshape((194,1))}
df = pd.DataFrame(station['swpies'],index=data_range,columns=[ridetype[0]])
for i in range(22):
    station={'swpies':mta_allsum[i+1].reshape((194,1))}
    df[ridetype[i+1]] = pd.DataFrame(station['swpies'],index=data_range)
```

```python
len(df)
```

```python
df.head()
```

```python
print df[ridetype[0]].mean(), df[ridetype[0]].std()
thresholds = [df[ridetype[0]].mean() + 3*df[ridetype[0]].std(), 
              df[ridetype[0]].mean() - 3*df[ridetype[0]].std()]
print thresholds
print min(df[ridetype[0]].index)
```

```python
f = plt.figure(figsize=(20,15))
for i in range(23):
    thresholds = [df[ridetype[i]].mean() + 3*df[ridetype[i]].std(), 
                  df[ridetype[i]].mean() - 3*df[ridetype[i]].std()]
    ax = plt.subplot(8,3,i+1)
    p = df[ridetype[i]].plot(logy=True)
    plt.xlabel(ridetype[i])
    p.plot([p.get_xlim()[0], p.get_xlim()[1]], [thresholds[1], thresholds[1]])
```

```python
for i in range(23):
    thresholds = [df[ridetype[i]].mean() + 3*df[ridetype[i]].std(), 
                  df[ridetype[i]].mean() - 3*df[ridetype[i]].std()]
    print df[ridetype[i]].loc[(df[ridetype[i]]<thresholds[1])]
```

#Task 2: 
##Some of the time series are stationary, some have a complex structure, some show a downward trend: Identify the **ridership types**  that have steadily increased in popularity,  and that have steadily decreased. by how much? (e.g what is the ratio of usage in the first 10 and last 10 weeks)

```python
ratio = np.zeros(23)
std = np.zeros(23)
fig = plt.figure(figsize=(20,20))
for i in range(23):
    ax = fig.add_subplot(8,3,i+1)
    rolling_mean = pd.rolling_mean(df[ridetype[i]], 10)
    ratio[i] = rolling_mean[9]/rolling_mean[193]
    std[i] = np.nanstd(rolling_mean)
    rolling_mean.plot()
    ax.set_title(ridetype[i])
```

```python
ratio_std = pd.DataFrame({'Type' : ridetype,'Flux Ratio From Beginning to End' : ratio,'Std' : std})
```

```python
ratio_std
```

##Comment:
- Increase: 14d, 7d, mc, rr, sem, spec
- Decrease: month, rfm
- Closed: afas, exp, ez

#Task 3:
##Several stations show a prominent annual periodicity. Identify the 4 **stations** (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). (Can you figure out what the periodic peak in rides is due to?)

```python
mta_allsum1 = mta_allsum_1.T
```

```python
df1 = pd.DataFrame(mta_allsum1)
```

```python
N = len(df1)

fig=plt.figure(figsize=(15,5))
for i in range(599):
    f = np.abs(np.fft.rfft(df1[i+1]))
    ax=fig.add_subplot(111)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), 'o', ms=5)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-', lw=.5);
ax.set_xticklabels([ "{:.1f}".format(1/x) for x in ax.get_xticks()], fontsize=12)
plt.xlabel("period (week)", fontsize=20)
plt.ylabel("Power", fontsize=20)
plt.axvline(1.0/52,lw=3, color='lightblue', label='at 1/52 weeks',alpha=0.8)
plt.legend()
```

```python
station = []
power = []
for i in range(599):
    x = (np.abs(np.fft.rfft(df1[i],)[1:]))
    station.append(i)
    power.append(x[3])
```

```python
dic = {'station':station,'power':power}
df2 = pd.DataFrame(dic)
```

```python
df2.sort_values('power', ascending=0).head(4)
```

Station 0,150,194 and 327 have the most prominent annual periodicity.
